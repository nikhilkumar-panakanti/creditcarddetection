```python
print("Outlier Analysis:")
print("="*80)
numerical_features = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'PAY_AMT1']
outlier_summary = []

for col in numerical_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_summary.append({
        'Feature': col,
        'Outliers': len(outliers),
        'Percentage': f"{len(outliers)/len(df)*100:.2f}%",
        'Lower Bound': f"{lower_bound:.2f}",
        'Upper Bound': f"{upper_bound:.2f}"
    })

outlier_df = pd.DataFrame(outlier_summary)
print(outlier_df.to_string(index=False))

fig, axes = plt.subplots(1, 4, figsize=(20, 5))
for idx, col in enumerate(numerical_features):
    axes[idx].boxplot(df[col], vert=True, patch_artist=True, 
                      boxprops=dict(facecolor='skyblue', alpha=0.7))
    axes[idx].set_ylabel('Value')
    axes[idx].set_title(col, fontsize=12, fontweight='bold')
    axes[idx].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## 7. Outlier Detection and Analysis

### 7.1 Outlier Identification Using IQR Method
Detecting and quantifying outliers in key numerical features.

## ANALYSIS: Outlier Detection

### 1. OUTLIER DEFINITION:
Using the **Interquartile Range (IQR) method**:
- Q1 = 25th percentile
- Q3 = 75th percentile
- IQR = Q3 - Q1
- Lower bound = Q1 - 1.5 × IQR
- Upper bound = Q3 + 1.5 × IQR
- Values outside these bounds are considered outliers

### 2. KEY OUTLIER FINDINGS:

```python
# Train all models with cross-validation
print("Training and Cross-Validating All Models")
print("="*80)

cv_scores = {}
all_results = {}

for model_name, model in models.items():
    print(f"\nTraining {model_name}...")
    
    # Use tuned model if available
    if model_name in tuned_models:
        model = tuned_models[model_name]
        print(f"  Using optimized hyperparameters")
    
    # Cross-validation on training data
    cv_f1 = cross_val_score(model, X_train_smote, y_train_smote, 
                            cv=cv_strategy, scoring='f1', n_jobs=-1)
    cv_scores[model_name] = cv_f1
    
    # Train on full training set
    start_time = time.time()
    model.fit(X_train_smote, y_train_smote)
    training_time = time.time() - start_time
    
    # Predictions
    y_pred = model.predict(X_test_scaled)
    y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]
    
    # Calculate metrics
    all_results[model_name] = {
        'model': model,
        'y_pred': y_pred,
        'y_pred_proba': y_pred_proba,
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred),
        'f1': f1_score(y_test, y_pred),
        'roc_auc': roc_auc_score(y_test, y_pred_proba),
        'cv_f1_mean': cv_f1.mean(),
        'cv_f1_std': cv_f1.std(),
        'training_time': training_time,
        'confusion_matrix': confusion_matrix(y_test, y_pred)
    }
    
    print(f"  CV F1-Score: {cv_f1.mean():.4f} (±{cv_f1.std():.4f})")
    print(f"  Test F1-Score: {all_results[model_name]['f1']:.4f}")
    print(f"  Training Time: {training_time:.2f}s")

# Create results DataFrame
results_data = []
for model_name, results in all_results.items():
    results_data.append({
        'Model': model_name,
        'Test Accuracy': results['accuracy'],
        'Test Precision': results['precision'],
        'Test Recall': results['recall'],
        'Test F1': results['f1'],
        'Test ROC-AUC': results['roc_auc'],
        'CV F1 (Mean)': results['cv_f1_mean'],
        'CV F1 (Std)': results['cv_f1_std'],
        'Training Time (s)': results['training_time']
    })
results_df = pd.DataFrame(results_data).sort_values('Test F1', ascending=False)

# Identify best model
best_model_name = results_df.iloc[0]['Model']

# Create cost DataFrame
cost_data = []
for model_name, results in all_results.items():
    cm = results['confusion_matrix']
    tn, fp, fn, tp = cm.ravel()
    
    # Cost calculation (example: FN costs 5x more than FP)
    cost_fn = 5  # Cost per false negative
    cost_fp = 1  # Cost per false positive
    total_cost = (fn * cost_fn) + (fp * cost_fp)
    avg_cost = total_cost / len(y_test)
    
    cost_data.append({
        'Model': model_name,
        'TN': tn,
        'FP': fp,
        'FN': fn,
        'TP': tp,
        'Total Cost': total_cost,
        'Avg Cost/Customer': avg_cost
    })
cost_df = pd.DataFrame(cost_data)

# Store best_params from tuned_models
best_params = {}
for model_name in tuned_models:
    best_params[model_name] = tuned_models[model_name].get_params()

print("\n" + "="*80)
print("✓ All models trained and validated")
```
